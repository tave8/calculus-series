```python
"""
[it works, but still looking to optimize]

Find N biggest or smallest numbers in a list, with a given order (ascending / descending), 
without sorting the list, without recursion and without any built-in method.


Notes:
    The separation of logic between smallest and biggest
    has been done for maintainability reasons.

    I'm aware some logic might feel duplicate, however with the priority
    abundance of readibility. 


 Giuseppe Tavella
"""

import math


def findNNumbers(n, how, array, order="asc"):

    # compute the first biggest number, whose upper limit is +infinity
    ret = []
    
    if how == "biggest":

        currBiggest = float("inf")
        
        for _ in range(n):

            nextBiggestButSmaller = findBiggestSmallerThan(currBiggest, array)
            # insert the next biggest in front of list
            if order == "asc":
                ret.insert(0, nextBiggestButSmaller)
            else:
                ret.append(nextBiggestButSmaller)

            currBiggest = nextBiggestButSmaller

    elif how == "smallest":

        currSmallest = float("-inf")
        
        for _ in range(n):

            nextSmallestButBigger = findSmallestBiggerThan(currSmallest, array)
            if order == "asc":
                ret.insert(0, nextSmallestButBigger)
            else:
                ret.append(nextSmallestButBigger)

            currSmallest = nextSmallestButBigger

    return ret



def findBiggestSmallerThan(upperLimit, array):
    lastBiggest = -math.inf
    
    for num in array:
        # strictly less than upperLimit means, 
        # that we are skipping the given value that represents
        # upper limit
        if num > lastBiggest and num < upperLimit:
            lastBiggest = num
    
    # edge case: if there is not enough "variety" of numbers, 
    # and still it's infinity, return the first number,
    # which was the same number in this array, all along
    if lastBiggest == -math.inf:
        return array[0]
    
    return lastBiggest


def findSmallestBiggerThan(lowerLimit, array):
    lastSmallest = math.inf
    
    for num in array:
        if num < lastSmallest and num > lowerLimit:
            lastSmallest = num
    
    if lastSmallest == math.inf:
        return array[0]
    
    return lastSmallest
            
            
            
findNNumbers(5, "smallest", [4, 2, 6, 3, -5, 35, 12], order="asc")
```
